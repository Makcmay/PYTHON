```python
from pylab import rcParams
rcParams ['figure.figsize'] = 20, 10
```

```python
grass_data = df[df['Type 1'] == 'Grass']
width = 0.2

fig, ax = plt.subplots()
rect = ax.bar(
    grass_data.Name, grass_data.Defense, width, label='Grass'
)

ax.set_ylabel('Условные единицы')
ax.set_xlabel('Имя')
ax.set_title('Защита')
ax.legend()

ax.bar_label(rect, padding=3)

fig.tight_layout()

plt.show()
```

```python
df.plot.scatter("Attack", "Defense", s=5);
```

```python
# по типу
sb.lmplot("Attack", "Defense", hue="Type 1", data=df, fit_reg=False);
```

```python
# Меняем цветовую палитру

# Получить все типы
types = df["Type 1"].unique()
# Получить цвет для каждого типа
cs = matplotlib.cm.tab20(np.linspace(0, 1, len(types)))

fig, ax = plt.subplots(figsize=(12, 6))
for c, t in zip(cs, types):
    df2 = df[df["Type 1"] == t]
    ax.scatter(df2["Attack"], df2["Defense"], color=c, label=t)
ax.legend(), ax.set_xlabel("Attack"), ax.set_ylabel("Defense");
```
